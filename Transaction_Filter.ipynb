```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',100)

trans_1_df = pd.read_csv("trans_1.csv",parse_dates=[2])
trans_2_df = pd.read_csv("trans_2.csv",parse_dates=[2])

loans_df = pd.read_csv("loan.csv",parse_dates=[2])
loans_df = loans_df.rename(columns={"date": "date_loan", "amount": "amount_loan"})

trans_1_df['account_id'] = trans_1_df['account_id'].astype('int64')
trans_1_df_temp = pd.merge(trans_1_df, loans_df, on='account_id', how='inner')
```

```python
trans_1_df_filter = trans_1_df_temp[(trans_1_df_temp['date']<trans_1_df_temp['date_loan'])]
trans_1_df_filter = trans_1_df_filter.drop(columns = ['loan_id','date_loan','amount_loan','duration','payments','status'])
```

```python
print(trans_1_df.shape)
print(trans_1_df_filter.shape)

trans_1_df_filter.head()
```

```python
trans_2_df['account_id'] = trans_2_df['account_id'].astype('int64')
trans_2_df_temp = pd.merge(trans_2_df, loans_df, on='account_id', how='inner')

trans_2_df_filter = trans_2_df_temp[(trans_2_df_temp['date']<trans_2_df_temp['date_loan'])]
trans_2_df_filter = trans_2_df_filter.drop(columns = ['loan_id','date_loan','amount_loan','duration','payments','status'])
```

```python
print(trans_2_df.shape)
print(trans_2_df_filter.shape)

trans_2_df_filter.head()
```

```python
trans_1_df_filter.to_csv('trans_1_filter.csv',index=False)

trans_2_df_filter.to_csv('trans_2_filter.csv',index=False)
```
